```
%%writefile local_modules/cooking_inst_mod/__init__.py

def get_precision_recall_f1(p, r): 
  index_pred = []
  index_real = []

  prediction_meals = []
  real_meals = []

  for i in range(0, len(p)-1):
    if p[i] == 1:
      index_pred.append(i)
      if (p[i+1] == 0): 
        if(len(index_pred)>5):        
          prediction_meals.append(index_pred)
        index_pred = []
    if r[i] == 1:
      index_real.append(i)
      if (r[i+1] == 0): 
        real_meals.append(index_real)
        index_real = []

  intersection = []
  tp = 0
  fp = 0

  for i in range(0, len(real_meals)): 
    intersection.append(0)

  for i in range(0, len(prediction_meals)): 
    flag_found = 0 
    count_int = 0

    for j in range(0, len(real_meals)):   
      flag_visited = 0

      for x in prediction_meals[i]:

        if x in real_meals[j]:      

          if intersection[j]==0:

            if count_int == 0: 
              count_int = count_int + 1
              intersection[j] = 1
              flag_visited = 1 
              flag_found = 1 
              tp=tp+1

            elif count_int > 0:          
              intersection[j] = 1
              flag_visited = 1
              flag_found = 1            
              fp=fp+1
              tp=tp+1

          elif flag_visited == 0:
                                 
            flag_found = 1
            flag_visited = 1

    if flag_found == 0: 
      fp=fp+1
    
    

  fn = 0

  for i in range(0, len(real_meals)):
    j=0
    flag_found = 0
    while j in range(0, len(prediction_meals)) and flag_found==0:

      for x in real_meals[i]:

        if x in prediction_meals[j]:    
          flag_found = 1

      j = j + 1  

    if flag_found == 0:
      fn=fn+1


  print("N° pasti reali:", len(real_meals))
  print("N° pasti predetti:", len(prediction_meals))

  print("TP:", tp)
  print("FP:", fp)
  print("FN:", fn)

  somma = tp+fp
  precision = (tp/(tp+fp))*100
  TPR = (tp/(tp+fn))*100
  F1 = 2*((precision*TPR)/(precision+TPR))
  print("Recall: %.2f %%" % TPR)
  print("Precision: %.2f %%" % precision)
  print("F1 score: %.2f %%" % F1)
```
